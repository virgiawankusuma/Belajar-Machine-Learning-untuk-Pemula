```python
# import pandas dan membuat dataframe
import pandas as pd
# dataframe = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
dataframe = pd.read_csv('imdb_indonesian_movies_2.csv')
```

```python
# drop kolom yang tidak dibutuhkan
dataframe = dataframe.drop(columns=
                           ['Unnamed: 0',
                            'Clothing ID', 
                            'Age', 
                            'Title', 
                            'Rating', 
                            'Recommended IND', 
                            'Positive Feedback Count',
                            'Division Name',
                            'Department Name'
                            ]
                           )
```

```python
# rename agar tidak ada spasi
# dataframe = dataframe.rename(columns={'Class Name' : 'Classname'})
dataframe = dataframe.drop(columns=['judul_film'])
# one hot encoding
category = pd.get_dummies(dataframe.genre)
newdataframe = pd.concat([dataframe, category], axis=1)
newdataframe = newdataframe.drop(columns=['genre'])
# newdataframe = newdataframe.drop(columns=['Classname'])
```

```python
# mengubah nilai dari dataframe ke numpy array
reviewtext = newdataframe['ringkasan_sinopsis'].values
label = newdataframe[['Drama', 'Horor','Komedi', 'Laga', 'Romantis']].values
# label = newdataframe[newdataframe.columns[1:10]].values
```

```python
# split data training dan testing
from sklearn.model_selection import train_test_split
reviewtext_train, reviewtext_test, label_train, label_test = train_test_split(reviewtext, label, test_size=0.2)
```

```python
# import tokenizer & pad sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
 
# mengubah kata ke bilangan numerik
tokenizer = Tokenizer(num_words=5000, oov_token='x')
tokenizer.fit_on_texts(reviewtext_train) 
tokenizer.fit_on_texts(reviewtext_test)
 
# convert menjadi sequence
sequence_train = tokenizer.texts_to_sequences(reviewtext_train)
sequence_test = tokenizer.texts_to_sequences(reviewtext_test)
padded_train = pad_sequences(sequence_train) 
padded_test = pad_sequences(sequence_test)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('bbc-news-data.csv', sep='\t')
```

```python
df.drop(['filename', 'content'], axis=1, inplace=True)

df.info()
```

```python
category = pd.get_dummies(df.category)
new_df = pd.concat([df, category], axis=1)
new_df = new_df.drop(columns=['category'])

new_df.info()
```

```python
news = new_df['title'].values
labels = new_df[['business', 'entertainment', 'politics', 'sport', 'tech']].values
```

```python
from sklearn.model_selection import train_test_split
news_train, news_test, labels_train, labels_test = train_test_split(news, labels, test_size=0.2)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=10000, oov_token='x', filters='!"#$&(1234567890),.-/;:~_')
tokenizer.fit_on_texts(news_train) 
tokenizer.fit_on_texts(news_test)
 
sequence_training = tokenizer.texts_to_sequences(news_train)
sequence_testing = tokenizer.texts_to_sequences(news_test)
 
padded_training = pad_sequences(sequence_training) 
padded_testing = pad_sequences(sequence_testing)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from keras import regularizers
from tensorflow.keras import layers

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=10000, output_dim=16),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dropout(0.6),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax'),
])

opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

```python
class callback(tf.keras.callbacks.Callback) :
  def on_epoch_end(self, epoch, logs={}) :
    if(logs.get('accuracy') > 0.9 and logs.get('val_accuracy') > 0.9) :
      self.model.stop_training = True
```

```python
num_epochs = 1000

history = model.fit(padded_training, labels_train, batch_size=128, epochs=num_epochs,
                    validation_data=(padded_testing, labels_test), verbose=2, callbacks=callback())
```
