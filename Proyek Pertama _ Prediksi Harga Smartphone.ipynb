```python
dataframe = pd.read_csv('Flipkart_output.csv')
dataframe
```

```python
dataframe = dataframe.drop(columns=['ImageUrl',
                                    'Screen_size',
                                    'Camera_details'])
dataframe
```

```python
# dataframe['Battery_details'] = dataframe.Battery_details.str.replace(' mAh Battery', '')
# dataframe['Battery_details'] = dataframe.Battery_details.str.replace(' mAh Lithium-ion Battery', '')
# dataframe['Battery_details'] = dataframe.Battery_details.str.replace(' mAh Li-polymer Battery', '')

data_battery= dataframe['Battery_details'].str.split('(\d+)',  expand=True)

#Split kolom memori menjadi atribut memori dan tipe penyimpanan, biar mudah dalam melihat data
split_memori = dataframe['Storage_details'].str.split('|', 1, expand=True)
split_memori
#Inisiasi kolom baru yaitu tipe penyimpanan

dataframe['Baterai (mAh)'] = pd.to_numeric(data_battery[1])
dataframe['RAM (GB)'] = split_memori[0]
dataframe['ROM (GB)'] = split_memori[1]

dataframe = dataframe.drop(columns=['Battery_details',
                                    'Storage_details'])

dataframe
```

```python
dataframe.info()
```

```python
data_ram= dataframe['RAM (GB)'].str.split('(\d+)',  expand=True)
data_rom= dataframe['ROM (GB)'].str.split('(\d+)',  expand=True)
dataframe['RAM (GB)'] = pd.to_numeric(data_ram[1], downcast='integer')
dataframe['ROM (GB)'] = pd.to_numeric(data_rom[1], downcast='integer')
dataframe
```

```python
loc_data=dataframe.loc[(dataframe['Baterai (mAh)'] >= 3000)|
                       (dataframe['RAM (GB)'] >= 3)|
                       (dataframe['ROM (GB)'] >= 32)
                       ]
```
