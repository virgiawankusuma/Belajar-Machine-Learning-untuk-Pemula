```python
import pandas as pd
dataframe = pd.read_csv('citrus.csv')
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
dataframe.name[dataframe.name == 'orange'] = 0
dataframe.name[dataframe.name == 'grapefruit'] = 1
```

```python
dataset = dataframe.values
```

```python
dataset
```

```python
# pilih 4 kolom terakhir sebagai atribut
attribute = dataset[:,1:6]
# bilangan sebelum koma untuk memilih baris pada dataframe
# bilangan setelah koma untuk memilih kolom pada dataframe
```

```python
label = dataset[:,0]
```

```python
# Normalization
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
attribute_scale = min_max_scaler.fit_transform(attribute)
attribute_scale
```

```python
# pisahkan data training dan testing
from sklearn.model_selection import train_test_split
attribute_train, attribute_test, label_train, label_test = train_test_split(attribute_scale, label, test_size=0.3)
```

```python
import numpy as np

label_train = label_train.astype(np.float32)
label_test = label_test.astype(np.float32)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([    
                    Dense(32, activation='relu', input_shape=(5,)),    
                    Dense(32, activation='relu'),    
                    Dense(1, activation='sigmoid'),])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(attribute_train, label_train, epochs=100)
```

```python
model.evaluate(attribute_test, label_test)
# elemen pertama adalah loss dan elemen kedua adalah akurasi
```
