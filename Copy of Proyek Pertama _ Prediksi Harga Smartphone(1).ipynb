```python
dataframe = pd.read_csv('phone_dataset .csv', error_bad_lines=False)
dataframe
```

```python
dataframe = dataframe.drop(columns=['network_technology',	
                                    '2G_bands',
                                    '3G_bands',	
                                    '4G_bands',
                                    'network_speed',
                                    'GPRS',
                                    'EDGE',
                                    'announced',
                                    'status',
                                    'dimentions',
                                    'weight_g',
                                    'weight_oz',
                                    'SIM',
                                    'display_type',
                                    'display_resolution',
                                    'display_size',
                                    'GPU',
                                    'memory_card',
                                    'primary_camera',
                                    'secondary_camera',
                                    'loud_speaker',
                                    'audio_jack',
                                    'WLAN',
                                    'bluetooth',
                                    'GPS',
                                    'NFC',
                                    'radio',
                                    'USB',
                                    'sensors',
                                    'colors',
                                    'img_url'
                                    ])
dataframe
```

```python
os= dataframe['OS'].str.split('(\d+)',  expand=True)
battery= dataframe['battery'].str.split('(\d+)',  expand=True)
# cpu= dataframe['CPU'].str.split('(\d+)',  expand=True)
ram= dataframe['RAM'].str.split('(\d+)',  expand=True)
rom= dataframe['internal_memory'].str.split('(\d+)',  expand=True)

dataframe['Android Version'] = pd.to_numeric(os[1])
dataframe['Battery (mAh)'] = pd.to_numeric(battery[1])
# dataframe['CPU_Clockspeed (GHz)'] = pd.to_numeric(cpu[1])
dataframe['RAM (GB)'] = pd.to_numeric(ram[1])
dataframe['ROM (GB)'] = pd.to_numeric(rom[1])

dataframe = dataframe.drop(columns=['OS'])
dataframe = dataframe.drop(columns=['battery'])
# dataframe = dataframe.drop(columns=['CPU'])
dataframe = dataframe.drop(columns=['RAM'])
dataframe = dataframe.drop(columns=['internal_memory'])

dataframe
```

```python
dataframe.dropna(subset = [
                           'brand',
                           'model',
                           'CPU',
                           'Chipset',
                           'approx_price_EUR',
                           'Android Version',
                           'Battery (mAh)',
                           'RAM (GB)',
                           'ROM (GB)'
                           ],
                 inplace=True)
dataframe
```
